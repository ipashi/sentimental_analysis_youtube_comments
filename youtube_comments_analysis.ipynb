_removing out the emoji's using unicode values of emoji's_

```python
import re
ytwithout = open("yt_comments_without.csv","w",encoding="utf-8")
with open("yt_comments.csv","r",encoding="UTF-8") as yt:
    for line in yt:
        line = line.strip()
        emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
        print(emoji_pattern.sub(r"",line),sep=",",file = ytwithout)
ytwithout.close()
yt.close()
```

```python
import re
with open("yt_comments_without.csv","r",encoding="utf-8") as ytw:
    for line in ytw:
        tag_pattern = re.compile("<\w>|</\w>")
        print(tag_pattern.sub(r"",line),sep=",")
ytw.close()
```

_removing anchor tags of urls_

```python
import re
yttag = open("yt_comments_tagsout.csv","w",encoding="utf-8")
with open("yt_comments_without.csv","r",encoding="utf-8") as ytw:
    for line in ytw:
        tag_pattern1 = re.compile("<[A-Za-z][a-zA-Z]*>|<[A-Za-z][a-zA-Z]\s/>*|</[A-Za-z]*>|<a href=\"https:\/\/www.youtube.com\/watch\?v=[\w]*&\w*;\w=\w*\">")
        print(tag_pattern1.sub(r"",line),file=yttag,sep=",")
yttag.close()
ytw.close()
```

_removing special symbols_

```python
import re
ytspec = open("yt_comments_withoutspec.csv","w",encoding="utf-8")
with open("yt_comments_tagsout.csv","r",encoding="utf-8") as ytw:
    for line in ytw:
        tag_pattern = re.compile("[\d?!*&#@$%()-+.;:,]|\n*")
        print(tag_pattern.sub(r"",line),file=ytspec,sep=",")
ytspec.close()
ytw.close()
```


__sentimental analysis will output the values of polarity and subjectivity
  subjectivity has nothing to do with the sentimental analysis
  only polarity values needed do the check
   positive values indicate the good feedback
   negative values indicate the bad feedback__
   

```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
from textblob import TextBlob
stop_words = set(stopwords.words("english"))
count = 0
pol = 0.0
with open("yt_comments_withoutspec.csv","r",encoding="utf-8") as rd:
    for line in rd:
        line = line.lower()
        if (len(line)) != 1:
            words = word_tokenize(line)
            without_stop = []
            for i in words:
                if i not in stop_words:
                    without_stop.append(i)
                    without_stop_string = " ".join(without_stop)
                    #print(without_stop_string)
                    analysis = TextBlob(without_stop_string)
                    #print(analysis.sentiment)
                    if analysis.sentiment[0] != 0.0 and analysis.sentiment[1] != 0.0:
                        count = count+1
                        pol = pol + analysis.sentiment[0]
print(pol/count)
```
